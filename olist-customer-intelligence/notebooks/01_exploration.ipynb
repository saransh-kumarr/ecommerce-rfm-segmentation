```python
import pandas as pd

customers = pd.read_csv('../data/olist_customers_dataset.csv')
orders = pd.read_csv('../data/olist_orders_dataset.csv')
payments = pd.read_csv('../data/olist_order_payments_dataset.csv')
items = pd.read_csv('../data/olist_order_items_dataset.csv')
products = pd.read_csv('../data/olist_products_dataset.csv')
sellers = pd.read_csv('../data/olist_sellers_dataset.csv')
reviews = pd.read_csv('../data/olist_order_reviews_dataset.csv')
geo = pd.read_csv('../data/olist_geolocation_dataset.csv')
category = pd.read_csv('../data/product_category_name_translation.csv')

print("all loaded")
```

```python
for name, df in [('customers', customers), ('orders', orders), ('payments', payments),
                  ('items', items), ('products', products), ('sellers', sellers),
                  ('reviews', reviews), ('geo', geo), ('category', category)]:
    print(f"\n--- {name} ---")
    print("Shape:", df.shape)
    print(df.isnull().sum())
```

```python
# see first few rows of the most important tables
print(f"orders \n {orders.head(3)} \n")
print(f"items \n {items.head(3)} \n")
print(f"payments \n {payments.head(3)} \n")

```

```python
# check order status types
print(orders['order_status'].value_counts())
```

```python
# check data types of main tables
for name, df in [('orders', orders), ('items', items), ('payments', payments), ('customers', customers)]:
    print(f"\n--- {name} ---")
    print(df.dtypes)
```

```python
# check duplicates in main tables
for name, df in [('orders', orders), ('items', items), ('payments', payments), ('customers', customers)]:
    print(f"{name} duplicates: {df.duplicated().sum()}")
```

```python
# check if order_id in items all exist in orders
items_orders_match = items['order_id'].isin(orders['order_id']).all()
print("All item order_ids exist in orders:", items_orders_match)

# check if order_id in payments all exist in orders
payments_orders_match = payments['order_id'].isin(orders['order_id']).all()
print("All payment order_ids exist in orders:", payments_orders_match)

# check if customer_id in orders all exist in customers
customers_match = orders['customer_id'].isin(customers['customer_id']).all()
print("All order customer_ids exist in customers:", customers_match)
```
